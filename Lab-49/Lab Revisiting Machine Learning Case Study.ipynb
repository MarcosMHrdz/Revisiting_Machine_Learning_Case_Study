Identificar las columnas con más de un 85% de valores perdidos

Elimine esas columnas del marco de datos

```python
# (df["PVASTATE"]==' ').sum()/df["PVASTATE"].count()
```

```python
df.replace(" ", 0, inplace=True)# Calcular la proporción de valores no nulos en la columna "PVASTATE"
proporcion_no_nulos = df[df["PVASTATE"] != 0]["PVASTATE"].count() / len(df)
print("Proporción de valores no nulos en PVASTATE después de reemplazar espacios en blanco por 0:", proporcion_no_nulos)
```

```python
for column in df.columns:
    # Calcular la proporción de valores diferentes de 0 en la columna actual
    proporcion_no_cero = df[df[column] != 0][column].count() / len(df)
    
    # Verificar si la proporción es mayor o igual al 15%
    if proporcion_no_cero <= 0.15:
        print(f"Eliminando columna '{column}' porque la proporción de valores diferentes de 0 es {proporcion_no_cero:.2f}")
        df.drop(column, axis=1, inplace=True)

# Mostrar las columnas restantes en el DataFrame
print(df.columns)
```
